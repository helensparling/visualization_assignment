```python
## Question 2.4:  kernel density plot, boxplot, and statistical description.
```

```python

#histogram
hist_plot = sns.histplot(college_df["grad_100_value"])
hist_plot.set(title='Percent of students that graduate within the expected timeframe',xlabel='grad_100_value',ylabel='Count')
plt.show()
```

```python
# Kernel density plot
kd_plot = sns.kdeplot(college_df["grad_100_value"])
kd_plot.set(title='Percent of students that graduate within the expected timeframe',xlabel = 'grad_100_value',ylabel='Count')
plt.show()
```

```python
# boxplot

box_plot = sns.boxplot(college_df["grad_100_value"])
box_plot.set(title='Percent of students that graduate within the expected timeframe',xlabel = 'grad_100_value',ylabel='Count')
plt.show()
```

```python
# Statistical description of grad_100_value variable:

descr = college_df['grad_100_value'].describe()
print(descr,'\n')

std = descr['std']
var = std**2
print('Variance: ', var,'\n')

thirdQ = descr['75%']
firstQ = descr['25%']
IQR = thirdQ - firstQ
print('IQR: ', IQR)
```

```python

sns.kdeplot( data=college_df, x="grad_100_value", hue="control", common_norm=True).set(title='Grad_100_value KDE Plot Grouped by Control')
# Public schools peak around a 15 grad_100_value, while private for-profit and private not-for profit have more spread out distributions (peaking 
# around 5 and 40, respecitvely)
```

```python
sns.kdeplot( data=college_df, x="grad_100_value", hue="level", common_norm=True).set(title='Grad_100_value KDE Plot Grouped by Level')
plt.show()
```

```python
# Use groupby and .describe to make grouped calculations of statistical descriptions of grad_100_value by level and control. Which 
# institutions appear to have the best graduation rates?
#college_df.loc[:,["level","grad_100_value"]].groupby("level").describe()
college_df.loc[:,['level','control','grad_100_value']].groupby(['level','control']).describe()
```

```python
# Based on these calculations, it appears that 2 year private not-for-profit institutions have the best graduation rates.
```

```python
## Question 2.6:

college_df['levelXcontrol']=college_df['level']+', '+college_df['control']

sns.kdeplot( data=college_df, x="grad_100_value", hue="levelXcontrol", common_norm=True).set(title='Grad_100_value KDE Plot Grouped by Level x Control')
plt.show()
```

```python
## Question 2.7: Make a scatterplot of grad_100_value by aid_value. Describe what you see. Now make the same plot, grouping by level and then control. 
# Describe what you see. For which kinds of institutions does aid seem to increase graduation rates?

college_df.plot.scatter(y='grad_100_value',x='aid_value').set(title='Graduation rates based on financial aid',xlabel= 'financial aid', ylabel = 'graduation rates')
plt.show()
# Most of the data in the scatterplot is clustered toward the bottom left. It seems that those who do not graduate within the expected time frame also
# tend to be the individuals who recieve less aid. 
```

```python
grouped_scatterplot = sns.scatterplot(data=college_df,x="aid_value", y="grad_100_value", hue ="levelXcontrol").set(title='Graduation rates grouped by level and control',xlabel= 'financial aid', ylabel = 'graduation rates')
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.show()
# This scatterplot shows 4-year, private not-for-profit institutions see the most increase in graduation rates as financial aid increases. Aid for 2 
# year private for-profit institutions seems to not vary much with graduation rates spread widely. 
```
